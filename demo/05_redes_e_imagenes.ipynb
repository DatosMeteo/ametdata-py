## 2. Radar regional

Productos de radar regional disponibles (ejemplos): `CAT`, `AND`, `LEV`, `VAL`, `GAL`, `CAN`, `BAL`...

```python
from aemetdata.redes import radar_regional

imagen_regional = await radar_regional("CAT", [API_KEY])  # Cataluna
print(f"Bytes descargados (radar CAT): {len(imagen_regional)}")

if imagen_regional:
    display(Image(data=imagen_regional))
```

## 3. Mapa de rayos

```python
from aemetdata.redes import rayos_mapa

resp_rayos = await rayos_mapa([API_KEY])
print("Respuesta API rayos:")
print(resp_rayos)

# Si la URL de datos es una imagen, descargarla y mostrarla
if resp_rayos.get("datos"):
    from aemetdata.utils.suport_functions import fetch_bytes_url
    img = await fetch_bytes_url(resp_rayos["datos"], "rayos")
    print(f"Imagen de rayos: {len(img)} bytes")
    if img:
        display(Image(data=img))
```

## 4. Contaminación de fondo

```python
from aemetdata.redes import contaminacion_fondo

# estacion y magnitud se consultan previamente en la web de AEMET
resp_cont = await contaminacion_fondo("CBA", "BEN", [API_KEY])
print("Contaminacion de fondo:")
print(resp_cont)
```

## 5. Ozono

```python
from aemetdata.redes import ozono, perfil_ozono

resp_ozono = await ozono("EL", [API_KEY])  # El Arenosillo
print("Datos de ozono:")
print(resp_ozono)
```

```python
resp_perfil = await perfil_ozono("EL", [API_KEY])
print("Perfil vertical de ozono:")
print(resp_perfil)
```

## 6. Radiación solar

```python
from aemetdata.redes import radiacion

resp_rad = await radiacion("SEV", "GHI", [API_KEY])  # Sevilla, Global Horizontal Irradiance
print("Radiacion solar (SEV, GHI):")
print(resp_rad)
```

## 7. Imágenes de satélite

```python
from aemetdata.imagenes import satelite_ndvi

# NDVI: Indice de Vegetacion de Diferencia Normalizada
img_ndvi = await satelite_ndvi([API_KEY])
print(f"NDVI descargado: {len(img_ndvi)} bytes")

if img_ndvi:
    display(Image(data=img_ndvi))
```

```python
from aemetdata.imagenes import satelite_sst

# SST: Sea Surface Temperature
img_sst = await satelite_sst([API_KEY])
print(f"SST descargado: {len(img_sst)} bytes")

if img_sst:
    display(Image(data=img_sst))
```
