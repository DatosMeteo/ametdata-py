## 1. Radar nacional

```python
from aemetdata.redes import radar_nacional
from aemetdata.utils.suport_functions import AemetError

try:
    imagen_radar = await radar_nacional([API_KEY])
    print(f"Bytes descargados: {len(imagen_radar)}")
    if imagen_radar:
        display(Image(data=imagen_radar))
except AemetError as e:
    print(f"Endpoint no disponible: {e}")
```

## 2. Radar regional

Productos de radar regional disponibles (ejemplos): `CAT`, `AND`, `LEV`, `VAL`, `GAL`, `CAN`, `BAL`...

```python
from aemetdata.redes import radar_regional

try:
    imagen_regional = await radar_regional("CAT", [API_KEY])  # Cataluna
    print(f"Bytes descargados (radar CAT): {len(imagen_regional)}")
    if imagen_regional:
        display(Image(data=imagen_regional))
except AemetError as e:
    print(f"Endpoint no disponible: {e}")
```

## 3. Mapa de rayos

```python
from aemetdata.redes import rayos_mapa

try:
    resp_rayos = await rayos_mapa([API_KEY])
    print("Respuesta API rayos:")
    print(resp_rayos)
    if resp_rayos.get("datos"):
        from aemetdata.utils.suport_functions import fetch_bytes_url
        img_rayos = await fetch_bytes_url(resp_rayos["datos"])
        if img_rayos:
            display(Image(data=img_rayos))
except AemetError as e:
    print(f"Endpoint no disponible: {e}")
```
