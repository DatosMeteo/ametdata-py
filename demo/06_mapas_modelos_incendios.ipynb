## 4. Modelos numéricos

### 4.1 AESPOL — Oleaje Atlántico Norte

Parámetros: `hs` (altura ola), `theta0` (dirección media), `tmm10` (periodo medio).

```python
from aemetdata.modelos import aespol

resp_aespol = await aespol(pasada="00", alcance="12", parametro="hs", api_keys=[API_KEY])
print("Modelo AESPOL (hs, pasada 00, alcance 12):")
print(resp_aespol)
```

### 4.2 AEWAM — Oleaje Atlántico

Parámetros: `shts`, `shww`, `swh`, `mdts`, `mdww`, `mwd`, `mpts`, `mpww`, `mwp`.

```python
from aemetdata.modelos import aewam

resp_aewam = await aewam(pasada="00", alcance="24", parametro="swh", api_keys=[API_KEY])
print("Modelo AEWAM (swh, pasada 00, alcance 24):")
print(resp_aewam)
```

### 4.3 Harmonie — Modelo numérico de alta resolución

Parámetros más habituales: `33` (viento), `61` (precipitación), `11` (temperatura 2m), `228` (humedad relativa).

Pasadas: `00`, `06`, `12`, `18`. Formatos: `grib` o `netcdf`.

```python
from aemetdata.modelos import harmonie_peninsula

resp_harm = await harmonie_peninsula(
    pasada="00", alcance="24", parametro="11", formato="grib", api_keys=[API_KEY]
)
print("Harmonie Peninsula (T2m, pasada 00, alcance 24, grib):")
print(resp_harm)
```

```python
from aemetdata.modelos import harmonie_canarias

resp_harm_can = await harmonie_canarias(
    pasada="12", alcance="48", parametro="33", formato="netcdf", api_keys=[API_KEY]
)
print("Harmonie Canarias (viento, pasada 12, alcance 48, netcdf):")
print(resp_harm_can)
```
