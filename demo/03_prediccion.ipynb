## 2. Predicción por CCAA

## 3. Predicción por provincia

## 4. Predicción específica por municipio

## 5. Predicción de playa

## 6. UVI (Índice Ultravioleta)

```python
from aemetdata.prediccion import uvi

# dia: '0' = hoy, '1' = manana, '2' = pasado, '3' = en 3 dias
pred_uvi = await uvi("0", [API_KEY])
print(f"Prediccion UVI hoy: {len(pred_uvi)} registros")
pd.DataFrame(pred_uvi).head(10)
```
